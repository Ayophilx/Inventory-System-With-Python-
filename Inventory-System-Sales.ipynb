```python
# STEP 1
import json
import time
"""load the product details Json into a dictionary"""
with open(r"productDetails.json","r+") as fd:
    js = fd.read()
prd_details = json.loads(js)

#STEP 2

"""Display available products to the user"""

print('-'*20,"MENU","-"*20)
for key in prd_details.keys():
    print(key,prd_details[key]["Name"],prd_details[key]["Amount"],prd_details[key]["Quantity"])
print("")
print("-"*50)

#STEP 3

"""Collect user details for sale record and the product id requested to calculate the bill"""


user_name     = input("Enter your Name: ")
user_mail     = input("Enter your Email: ")
user_ph       = input("Enter your Phone No: ")
user_pr       = str(input("Enter product ID : "))
user_qt       = int(input("Enter product Quantity : "))
print("-"*50)

if(prd_details[user_pr]["Quantity"]>=user_qt):
    print("Name            :  ", prd_details[user_pr]["Name"])
    print("Price (Naira)   :  ", prd_details[user_pr]["Amount"])
    print("Quantity        :  ", user_qt)
    print("-"*50)
    print("Billing         :  ",user_qt * prd_details[user_pr]["Amount"], "#")
    print("-"*50)
    
    """ updating inventory"""
    prd_details[user_pr]["Quantity"] = int(prd_details[user_pr]["Quantity"]) - user_qt
    
    """Generate and update sale information"""
    sale = user_name +","+ user_ph +"," + user_mail +","+user_pr + "," + prd_details[user_pr]["Name"] +","+ str(user_qt)+ ","+str(prd_details[user_pr]["Amount"])+","+str(user_qt * prd_details[user_pr]["Amount"]) +","+time.ctime()+ "\n"
    
    """save product details"""
    js = json.dumps(prd_details)
    with open('productDetails.json','w') as fd:
        fd.write(js)
    
    """save Sales Record"""

    with open('Sales.txt',"a") as fd:
        fd.write(sale)

    print("")
    print("-"*50)
    print("   Thanks for your order, Inventory Update!")
    print("-"*50)
    print(sale)

    # If there's no enough product in the inventory to satisfy the request   
else:
    print("\n\n")
    print("Sorry, Not enough quantity of product  available.")
    print("Available Quantity is "+ str(prd_details[user_pr]["Quantity"]))
    print("Would you like to purchase other quantity\n Press Y for yes\n Press N for No")
    print("-"*50)
    
    """Collect new quantity from user"""
    response = input()
    if(response.capitalize()=="Y"):
        newQt = int(input("Please enter quantity : "))
        if(newQt <= prd_details[user_pr]["Quantity"]):      # Making sure the new order can be processed
            print("Name            :  ", prd_details[user_pr]["Name"])
            print("Price (Naira)   :  ", prd_details[user_pr]["Amount"])
            print("Quantity        :  ", newQt)
            print("-"*50)
            print("Billing         :  ",newQt * prd_details[user_pr]["Amount"], "#")
            print("-"*50)
            
            """updating inventory"""
            prd_details[user_pr]["Quantity"] = int(prd_details[user_pr]["Quantity"]) - newQt
    
            """Generate and update sale information"""
            
            sale = user_name +","+ user_ph +"," + user_mail +","+user_pr + "," + prd_details[user_pr]["Name"] +","+ str(newQt)+ ","+str(prd_details[user_pr]["Amount"])+","+str(newQt * prd_details[user_pr]["Amount"]) +","+time.ctime()+ "\n"
            
            js = json.dumps(prd_details)
            with open('productDetails.json','w') as fd:
                fd.write(js)
    
            """save Sales Record"""

            with open('Sales.txt',"a") as fd:
                fd.write(sale)
            print("")

            print("-"*50)
            print("   Thanks for your order, Inventory Update!")
            print("-"*50)

            print(sale)
        
        else:
            print("Cannot process your order, quantity higher than currently available\nPlease order less!")
            


    else:
        print("Thank you!")
        


        
        


```
